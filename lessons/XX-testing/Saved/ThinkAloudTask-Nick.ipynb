## The Task

Your E-commerce store accepts payments using credit cards. You received a notice from Discover (a credit card company) that one of your customer's credit cards was used illegally. Discover is  asking for a list of customer emails for all pending orders so they can notify them of the activity.

Orders are here: `https://raw.githubusercontent.com/mafudge/datasets/refs/heads/master/orders/task.csv`

```python
import pandas as pd

res = pd.read_csv("https://raw.githubusercontent.com/mafudge/datasets/refs/heads/master/orders/task.csv")
```

```python
res
```

```python
filtered_df = res[(res['orderstatus'] == 'pending') & (res['ordercreditcard'] == 'Discover')]
```
