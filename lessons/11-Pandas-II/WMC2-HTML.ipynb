```python
import pandas as pd
```

```python
# HTTP Error - need to set a user agent
tables = pd.read_html("https://ist256.github.io")
```

```python
# a good default user agent is just Mozilla/5.0
# we set this in the headers using storage_options
headers = {"User-Agent": "Mozilla/5.0"}
tables = pd.read_html("https://ist256.github.io", storage_options=headers)
```

```python
# tables is a list... of every table on the page

len(tables)
```

```python
headers = {"User-Agent": "Mozilla/5.0"}
tables = pd.read_html("https://www.baseball-reference.com/leagues/MLB-standings.shtml", storage_options=headers)
len(tables)
```

```python
# loop over each table and display
from IPython.display import display
for index in range(len(tables)):
    print("INDEX:", index)
    display(tables[index])
```

```python
tables[0]["division"] = "AL/East"
tables[1]["division"] = "A/Central"
pd.concat(tables)
```

```python
tables[0]['League'] = 'AL'
tables[0]['Division'] = 'East'
```

```python
teams_df = pd.concat(tables, ignore_index=True)
```
